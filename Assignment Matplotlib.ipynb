## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

#### Note: Also add title, xlabel, and ylabel to the plot.

#### A scatter plot is a type of plot used to display the relationship or correlation between two numerical variables. It is a graphical representation that shows individual data points as markers on a graph.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is scatter plot")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
## Use the following data:


#### The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or charts in a single figure, organized in a grid layout with rows and columns.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10, 8))

#plot 1
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title('Line Plot 1')

#plot2
plt.subplot(2, 2, 2)
plt.plot(x2,y2)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title('Line Plot 2')

#plot 3
plt.subplot(2, 2, 3)
plt.plot(x3,y3)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title('Line Plot 3')

#plot 4
plt.subplot(2, 2, 4)
plt.plot(x4,y4)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


#### A bar plot, also known as a bar chart, is a graphical representation that displays categorical data using rectangular bars. Each bar represents a category, and the length or height of the bar corresponds to the value or frequency of the category.
#### Bar plots are used to visually compare and represent the distribution, composition, or comparison of different categories or groups. They are particularly useful when working with discrete or categorical data, where each category represents a distinct group or variable.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title("Companies Profit in Bar Plot")
```

```python
plt.barh(company, profit)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title("Companies Profit in Horizontal Bar Plot")
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


#### A box plot, also known as a box and whisker plot, is a graphical representation that displays the distribution of a dataset through five summary statistics: minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum. It provides a concise and visual summary of the data's central tendency, dispersion, and presence of outliers.
#### The box plot is used for several purposes due to its ability to provide a concise summary of the distribution of a dataset. for example Outlier detection, Visualize non-normality and skewness etc

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig, ax = plt.subplots()

boxplot = [box1,box2]
ax.boxplot(boxplot)
ax.set_xticklabels(['Box 1','Box 2'])
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
